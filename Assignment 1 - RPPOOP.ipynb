```python
# Base Classes
# reference image -> 
class Assembly:
    def __init__(self):
        self.type = "Low-Level"
        self.releaseYear = 1949
        self.description = "Programming language for communicating with the hardware directly"

class Assembler:
    def __init__(self):
        self.description = "Software that converts an assembly language code to machine code"
        self.type = "Software"
```

```python
# Multiple Inheritance

class C(Assembler, Assembly):
    def __init__(self):
        self.type = "Low-Level"
        self.releaseYear = 1972
        self.description = "General-purpose computer programming language"
    
    def getType(self):
        return self.type
    
    def getReleaseDate(self):
        return self.releaseDate
    
    def getDescription(self):
        return self.description
```

```python
# Multi Level Inheritance
# Single Inheritance

# Hybrid Inheritance
class Cpp(C):
    def __init__(self):
        self.type = "Low-Level"
        self.releaseYear = 1985
        self.description = "An object-oriented programming language which gives a clear structure to programs and allows code to be reused, lowering development costs"
    
    def getType(self):
        return self.type
    
    def getReleaseDate(self):
        return self.releaseDate
    
    def getDescription(self):
        return self.description
    
class Go(Cpp):
    def __init__(self):
        self.type = "High-Level"
        self.releaseYear = 2009
        self.description = "Statically typed, compiled programming language"
    
    def getType(self):
        return self.type
    
    def getReleaseDate(self):
        return self.releaseDate
    
    def getDescription(self):
        return self.description
    
    
```

```python
# Hierarchical Inheritance

class ObjectiveC(C):
    def __init__(self):
        self.type = "High-Level"
        self.releaseYear = 1980
        self.description = "Object oriented language with smalltalk style messaging in C language"
    
    def getType(self):
        return self.type
    
    def getReleaseDate(self):
        return self.releaseDate
    
    def getDescription(self):
        return self.description

class Java(C):
    def __init__(self):
        self.type = "High-Level"
        self.releaseYear = 1996
        self.description = "Class-based, object-oriented programming language that is designed to have as few implementation dependencies as possible."
    
    def getType(self):
        return self.type
    
    def getReleaseDate(self):
        return self.releaseDate
    
    def getDescription(self):
        return self.description
        
```
